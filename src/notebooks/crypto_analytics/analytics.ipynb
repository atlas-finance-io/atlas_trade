```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import requests
import datetime
plt.style.use('dark_background')
csv_path = '/Users/markwindsor/Desktop/atlas_trade/src/notebooks/crypto_analytics/top_coins.csv'
top_coins_df = pd.read_csv(csv_path)

ids = top_coins_df['id'].tolist()
ids

# Generate date range from January 1, 2017, to today's date
date_rng = pd.date_range(start='2017-01-01', end=datetime.datetime.now().strftime('%Y-%m-%d'), freq='D')

# Create DataFrame with 'date' as the index and 'marketcapsum' as a column
df = pd.DataFrame(date_rng, columns=['date'])
df['marketcapsum'] = 0 
df.set_index('date', inplace=True)

df.reset_index(inplace=True)


for index, row in top_coins_df.iterrows():
    id = row['id']
    name = row['name']

    url = f"https://91j3dag4m9.execute-api.us-east-1.amazonaws.com/master/getCMCPriceHistory/{id}?startDate=2017-01-01"
    response = requests.get(url)
    data = response.json()

    for entry in data["priceHistory"]:
        date_entry = pd.Timestamp(entry['d'])
        
        if date_entry in df['date'].values:
            existing_value = df.loc[df['date'] == date_entry, 'marketcapsum'].values[0]
            df.loc[df['date'] == date_entry, 'marketcapsum'] = existing_value + entry["mcap"]
        else:
            print(f"Date {date_entry} not found in DataFrame.")


result_path = 'mcap_sum.csv'
df.to_csv(result_path, index=False)




    










```
